# Ejercicio
Se desea predecir la nota de un alumno en función a la cantidad de horas a las semana que estudió para un exámen.

Para ello se tomó una muestra de 6 alumnos , cuyos resultados se muestran en el siguiente cuadro.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#path='notas.csv'
#df= pd.read_csv(path,sep=',')
#df.head(6)
```

```python
x= [3,6,8,2,1,6]#HORA DE ESTUDIO
y= [8,10,15,8,5,12]#NOTAS
#x=df.iloc[:,0]
#y=df.iloc[:,1]
plt.title('Regresión')
plt.xlabel('x-> HORAS DE ESTUDIO')
plt.ylabel('y->NOTAS')
plt.grid(True)
plt.plot(x,y,'o',linewidth=4,color=(0.2,0.1,0.4))
```

LA ECUACION ESTIMADA ES:

Y = 3.5 + 1.5x

```python
horas=int(input('INGRESE LAS HORAS DE ESTUDIO:'))
#CALCULAMOS EL VALOR.
x.append(horas)
funcion = 3.5+(1.5*(horas))
y.append(funcion)

#DIBUJAMOS LA RECTA.
li=np.array(range(-1,horas))
fu=lambda num: 3.5+(1.5*(li))
#GRAFICA
plt.title('Regresión Actualizada')
plt.xlabel('x-> HORAS DE ESTUDIO')
plt.ylabel('y->NOTAS')
plt.grid(True)
plt.plot(x,y,'o',linewidth=3,color=(0.2,0.1,0.4))
plt.plot(li,fu(li))

print('LA NOTA PARA SUS HORAS DE ESTUDIO ES:')
print(funcion)
```
